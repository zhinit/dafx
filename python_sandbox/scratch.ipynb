```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_rate = 44100
freq = 200
audio_length = 2.0
total_samples = int(sample_rate * audio_length)
```

```python
saw_samples = []
times = []
for i in range(total_samples):
    time = i / sample_rate
    times.append(i/sample_rate)
    
    phase = (time * freq) % 1
    sample = 1 - phase * 2
    saw_samples.append(sample)

plt.plot(times[:1000], saw_samples[:1000])
```
